```python
import pandas as pd
import re
```

```python
df = pd.read_csv('data/Mental-Health-Twitter.csv')
df.shape
```

```python
df = df.drop(columns=['post_id','post_created','user_id','followers','friends','favourites','statuses','retweets','label'], axis=0)
df
```

```python

df.columns = df.columns.str.strip().str.lower()
```

```python
def clean_text(text):
    text = re.sub(r'@\w+', '', str(text))       # Remove mentions
    text = re.sub(r'http\S+', '', text)         # Remove links
    return text.lower().strip()
```

```python
df['clean_text'] = df['post_text'].apply(clean_text)
```

```python
df.head()
```

```python
df.to_csv("data/cleaned_tweets.csv", index=False)
```

```python
import pandas as pd

# Load uploaded 'tweets_pulled.csv' and previously saved 'cleaned_tweets.csv'
tweets_pulled_path = "data/tweets_pulled.csv"
cleaned_tweets_path = "data/cleaned_tweets.csv"

# Read both CSVs
tweets_pulled_df = pd.read_csv(tweets_pulled_path)
try:
    cleaned_tweets_df = pd.read_csv(cleaned_tweets_path)
except FileNotFoundError:
    cleaned_tweets_df = pd.DataFrame(columns=['Unnamed: 0', 'post_text', 'clean_text'])

# Standardize columns in tweets_pulled_df to match 'cleaned_tweets.csv'
tweets_pulled_df.columns = tweets_pulled_df.columns.str.strip().str.lower()
tweets_pulled_df = tweets_pulled_df.rename(columns={
    'uncleaned': 'post_text',
    'cleaned': 'clean_text'
})

# Add Unnamed: 0 column if missing
if 'unnamed: 0' not in tweets_pulled_df.columns:
    tweets_pulled_df.insert(0, 'Unnamed: 0', range(len(cleaned_tweets_df), len(cleaned_tweets_df) + len(tweets_pulled_df)))

# Append and remove duplicates if any
combined_df = pd.concat([cleaned_tweets_df, tweets_pulled_df], ignore_index=True).drop_duplicates()

# Save the updated dataframe
combined_csv_path = "data/cleaned_tweets.csv"
combined_df.to_csv(combined_csv_path, index=False)

combined_df.head()
```
